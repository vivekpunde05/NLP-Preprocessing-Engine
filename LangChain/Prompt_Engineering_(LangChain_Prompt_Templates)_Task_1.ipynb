```python
from langchain.prompts import PromptTemplate
from langchain.prompts.chat import ChatPromptTemplate
```

**Task 1 - Replace Hardcoded Prompts**

```python
#Create a reusable PromptTemplate
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

#Function using PromptTemplate
def explain_topic(topic):
    prompt = explain_template.format(topic=topic)
    return prompt

#Test
print(explain_topic("Artificial Intelligence"))
```

**Task 2 - Multi-Input Prompt System**

```python
#Multi input prompt template
multi_prompt_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

#Function to generate prompt
def generate_multi_prompt(topic, audience, tone):
    prompt = multi_prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
    return prompt

#Test cases
print(generate_multi_prompt("AI", "beginners", "friendly"))
print(generate_multi_prompt("Python", "kids", "fun"))
print(generate_multi_prompt("Deep Learning", "engineers", "technical"))
```

**Task 3 – Prompt Variations Engine**

```python
#Teaching template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

#Interview template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

#Storytelling template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a short engaging story."
)

#Function to generate variations
def prompt_variations(topic):

    teaching_prompt = teaching_template.format(topic=topic)
    interview_prompt = interview_template.format(topic=topic)
    story_prompt = story_template.format(topic=topic)

    return teaching_prompt, interview_prompt, story_prompt

#Test
teach, interview, story = prompt_variations("Machine Learning")
print("Teaching Prompt:", teach)
print("Interview Prompt:", interview)
print("Story Prompt:", story)
```

**Task 4 – ChatPromptTemplate System**

```python
#Role definitions
roles = {
    "teacher": "You are a helpful teacher who explains concepts clearly.",
    "interviewer": "You are a technical interviewer asking challenging questions.",
    "motivator": "You are a motivational mentor who encourages learning."
}

#Function to create chat prompts
def create_chat_prompt(role, topic):
    system_message = roles.get(role)
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    messages = chat_template.format_messages(topic=topic)
    return messages

#Test
chat_messages = create_chat_prompt("teacher", "Neural Networks")
for message in chat_messages:
    print(message)
```

**Task 5 – Input Validation Layer**

```python
#Input validation function
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Defaulting to beginner.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Defaulting to formal.")
        tone = "formal"

    return audience, tone


#Test
audience, tone = validate_inputs("kids", "exciting")
print("Audience:", audience)
print("Tone:", tone)
```

**Task 6 – Prompt Generator App**

```python
#Style templates
style_templates = {

    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Teach {topic} to {audience} learners in a {tone} tone."
    ),

    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Create interview questions about {topic} for {audience} in a {tone} tone."
    ),

    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style."
    )
}


#Prompt generator function
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    template = style_templates.get(style)

    if template is None:
        raise ValueError("Invalid style selected")

    prompt = template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return prompt


#Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7 – Template Reusability Test**

```python
#Reusable template
reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Provide a simple explanation of {topic}."
)

topics = [
    "Artificial Intelligence",
    "Machine Learning",
    "Deep Learning",
    "Neural Networks",
    "Natural Language Processing"
]

#Run template with multiple inputs
for topic in topics:
    result = reusable_template.format(topic=topic)
    print(result)
```
