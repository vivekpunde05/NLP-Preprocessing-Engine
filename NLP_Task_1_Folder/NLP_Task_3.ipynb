# Assignment NLP – 3: Chatbot using Transformers

## Task 1: Model Loading

In this step, I load a pre-trained transformer model from Hugging Face.
I use DialoGPT, which is designed for conversational AI tasks.

```python
# installing transformers library
!pip install transformers

# importing required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# loading pretrained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
```

## Task 2: User Input Handling

In this step, the chatbot accepts input from the user through the console.
This allows interaction between the user and the chatbot.

```python
# taking input from user
user_input = input("You: ")
print("User entered:", user_input)
```

## Task 3: Response Generation

The chatbot generates responses using the transformer model.
The input text is converted into tokens and passed to the model to generate meaningful output.

```python
# encoding input
input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

# generating response
output = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

# decoding response
response = tokenizer.decode(output[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

print("Chatbot:", response)
```

## Task 4: Continuous Conversation

To simulate a real chatbot, we maintain conversation history and allow multiple interactions.

```python
print("Chatbot: Hello! I am your AI assistant. Type 'exit' to stop.")

chat_history_ids = None

while True:
    
    # take user input
    user_input = input("You: ")
    
    # exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day.")
        break
    
    # encode input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')
    
    # maintain conversation history
    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids
    
    # generate response
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id
    )
    
    # decode response
    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )
    
    print("Chatbot:", response)
```

## Task 5: Exit Condition

The chatbot stops execution when the user types "exit" or "quit".
This ensures controlled termination of the conversation.
