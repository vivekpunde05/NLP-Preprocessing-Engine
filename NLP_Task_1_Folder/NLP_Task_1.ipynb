```python
#Task 2: Advanced Preprocessing Function
# Import required library
import re

# Preprocessing function from Task 2
def preprocess_text(text):
    # Handle empty input
    if text is None or text.strip() == "":
        return [], ""

    # Convert text to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (example: sooooo -> so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters and emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize text
    tokens = text.split()

    # Remove very short tokens except "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Reconstruct cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
# test cases
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# preprocessing and ouput

for sentence in test_sentences:

    tokens, cleaned_sentence = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```

```python
#Task 3: Stress Testing
sample_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in sample_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print(f"Original: {sentence}")
    print(f"Tokens: {tokens}")
    print(f"Cleaned: {cleaned}\n")
```

```python
#Task 4: Token Analytics
for sentence in sample_sentences:
    tokens, cleaned = preprocess_text(sentence)
    if tokens:
        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        avg_token_len = sum(len(t) for t in tokens) / total_tokens
    else:
        total_tokens = unique_tokens = avg_token_len = 0
        
    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total_tokens}, Unique Tokens: {unique_tokens}, Avg Token Length: {avg_token_len:.2f}\n")
```

```python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = []
for sentence in sample_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:", counter.most_common(10))
print("Top 5 Least Frequent Words:", counter.most_common()[:-6:-1])
```

```python
#Task 6: Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example usage
pipeline_output = full_pipeline(sample_sentences)
print(pipeline_output)
```
